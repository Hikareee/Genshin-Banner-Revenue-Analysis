```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df= pd.read_excel("Genshin Impact Revenue.xlsx")
df.dropna(inplace=True)
data = df.iloc[:,[0,5,6,7,8,9,10]]

#Correlation Matrix
corr_matrix = data.corr() #Compute correlation of columns

sn.heatmap(corr_matrix, annot=True) #Create Heatmap

plt.title("Correlation Matrix", fontsize =20)
plt.show()
```
